```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/drive/MyDrive/Practica2/Salary_dataset.csv')
```

```python
# Extraer las columnas de YearsExperience y Salary del DataFrame
YearsExperience = df['YearsExperience'].values
Salary = df['Salary'].values
# Calcular las medias de YearsExperience y Salary
x = np.mean(YearsExperience)
y = np.mean(Salary)

# Calcular la pendiente (coeficiente, m) y la ordenada al origen (intercepto, b)
numerador = np.sum((YearsExperience - x) * (Salary - y))
denominador = np.sum((YearsExperience - x) ** 2)
m = numerador / denominador
b = y - m * x
```

```python
# Función para predecir el salario
def PredictSalary(experience):
    return m * experience + b

# Generar valores para la línea de regresión
x_values = np.linspace(min(YearsExperience), max(YearsExperience), 100)
y_values = PredictSalary(x_values)

# Crear la gráfica
plt.scatter(YearsExperience, Salary, color='blue', label='Salario')
plt.plot(x_values, y_values, color='red', linewidth=2, label='Regresión Lineal')
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.title('Regresión Lineal')
plt.legend()
plt.grid(True)

# Mostrar la gráfica
plt.show()
```

```python
#Predicción
NewYearsExperience = 11.0
PredictedSalary = PredictSalary(NewYearsExperience)
PredictedSalary
```

```python
# Crear la gráfica
plt.scatter(YearsExperience, Salary, color='blue', label='Datos reales')
plt.plot(x_values, y_values, color='red', linewidth=2, label='Regresión Lineal')
plt.scatter(NewYearsExperience, PredictedSalary, color='green', marker='x', s=100, label='Predicción para 11 años')
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.title('Regresión Lineal Simple con Predicción')
plt.legend()
plt.grid(True)

# Mostrar la gráfica
plt.show()
```
